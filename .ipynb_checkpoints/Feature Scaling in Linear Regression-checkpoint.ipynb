```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.array([1000,1500,1250,3000,2400])
x_2 = np.array([2,3,2,6,4])
x = np.array([x_1,x_2])
y   = np.array([200,300,250,400,450])
feature = ["Sq Feet","No of Bed Rooms"]
```

```python
fig,ax=  plt.subplots(1, 2, figsize=(12, 3), sharey=False)
for i in range(len(ax)):
    ax[i].scatter(x[i],y)
    ax[i].set_xlabel(feature[i])
ax[0].set_ylabel("Price")
plt.show()
```

```python
#Mean Normalization Method
x_1_Normalized = (np.subtract(x_1,np.mean(x_1)))/(np.max(x_1) - np.min(x_1))
print(x_1_Normalized)
x_2_Normalized = (np.subtract(x_2,np.mean(x_2)))/(np.max(x_2) - np.min(x_2))
print(x_2_Normalized)

x_normalized = np.array([x_1_Normalized,x_2_Normalized])

y_normalized = (np.subtract(y,np.mean(y)))/(np.max(y) - np.min(y))
print(y_normalized)
```

```python
fig,ax=  plt.subplots(1, 2, figsize=(12, 3), sharey=False)
for i in range(len(ax)):
    ax[i].scatter(x_normalized[i],y_normalized)
    ax[i].set_xlabel(feature[i])
ax[0].set_ylabel("Price")
plt.show()
```

```python
#Z_Score Normalization
x_1_Z_Normalized = (np.subtract(x_1,np.mean(x_1)))/(np.std(x_1))
print(x_1_Z_Normalized)
x_2_Z_Normalized = (np.subtract(x_2,np.mean(x_2)))/(np.std(x_2))
print(x_2_Z_Normalized)

x_Z_normalized = np.array([x_1_Z_Normalized,x_2_Z_Normalized])

y_Z_normalized = (np.subtract(y,np.mean(y)))/(np.std(y))
print(y_Z_normalized)
```

```python
fig,ax=  plt.subplots(1, 2, figsize=(12, 3), sharey=False)
for i in range(len(ax)):
    ax[i].scatter(x_Z_normalized[i],y_normalized)
    ax[i].set_xlabel(feature[i])
ax[0].set_ylabel("Price")
plt.show()
```
